```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import cm, colors
import xarray as xr
import seaborn as sns
from pymosaic import *
from pyflowtrace import *
```

```python
md = xr.open_dataset('md.nc')
exp = xr.open_dataset('exp.nc')
```

```python
df = exp[['b',
          'contagion',
          'edge_density',
          'landscape_shape_index',
          'mean_contributing_area',
          'mean_outlet_distance',
          'std_contributing_area',
          'std_outlet_distance',
          'number_burned',
          'mean_burnflow_distance']].to_dataframe().reset_index()
```

```python
fig,ax = plt.subplots(ncols=4,figsize=[18,7])
sns.violinplot(data=df,x='b',y='number_burned',ax=ax[0]);
sns.violinplot(data=df,x='b',y='contagion',ax=ax[1]);
sns.violinplot(data=df,x='b',y='mean_outlet_distance',ax=ax[2]);
sns.violinplot(data=df,x='b',y='mean_burnflow_distance',ax=ax[3]);
ax[0].set_title('NUMBER OF CELLS BURNED')
ax[1].set_title('CONTAGION')
ax[2].set_title('MEAN DISTANCE FROM OUTLET')
ax[3].set_title('MEAN BURNFLOW')
plt.tight_layout()
```

```python
sns.pairplot(data=df,vars=['contagion','mean_outlet_distance','mean_burnflow_distance'],hue='b',height=3.5);
plt.tight_layout()
```
